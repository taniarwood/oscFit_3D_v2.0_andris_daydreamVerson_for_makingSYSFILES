# oscFit: Sub-leading neutrino flux uncertainties

#### This notebook shows how sub-leading neutrino flux uncertainties are implemented in oscFit. These uncertainties are:

####  (i) vertical-to-horizontal ratio, and

#### (ii) neutrino-to-antineutrino ratio. 

#### The main reference is Phys.Rev. D74 (2006) 094009 [astro-ph/0611266]. There are several caveats, so please read carefully.

The paper is almost ten years old, and it uses the Bartol flux as starting point. This was the best reference found so far, so we use it nonetheless. 

The authors refer to about 11 knobs. Since we want to add this to a minimizer, we don't implement them. People did this for PISA. You can ask them if you're curious. In oscFit we parameterize the net effect instead. We decided to take Fig. 7 (nu/nubar as function of energy) and Fig. 9 (hor/vert ratio) and come up with a parameterization that accurately reproduces Figs. 7-9. 

In IceCube analyses the MUON (anti)neutrino component is the leading one, so we pay special attention to it. The variations on the ELECTRON (anti)neutrino component are divided in two: one is FULLY correlated with the muon component, and one that is uncorrelated. The NuE component is sub-dominant, and the correlated part is sufficient to describe variations. Feel free to test the impact of the uncorrelated part yourself.

###### Important! The uncertainties given in the paper are absolute numbers on the RATIOS. A deviation can increase or decrease the ratio, and this can be done by modifying either of the fluxes entering the ratio. All these scenarios are implemented here. For example, for modifying the NuMu/NuMuBar flux there are 2 parameters: 1 that modifies the "strength" of the change, and 1 that decides how the change is shared between Nu and NuBar.

##### Q: How much does the flux as fcn of angle changes if I change the ratio as function of energy? The answer is no. See below.

```python
# Obtaining the flux modification (only E dependence)
# This figure is obtained modifying the neutrino to antineutrino ratio by ONE sigma

fluxmod = np.zeros_like(data['ecenters'])
for ei, energy in enumerate(data['ecenters']):
    fluxmod[ei] = 1+sf.LogLogParam(energy,e1max_mu, 
                                   e2max_mu, x1e, x2e)

# Need to multiply times the step-size in GeV to get the appropriate contribution
mu_newflux = data['numu_flux'].T*esteps*fluxmod
mu_oldflux = data['numu_flux'].T*esteps

lowe_bool = (data['ecenters']>0.3) * (data['ecenters']<3)
change_low = (np.sum(mu_newflux[:,lowe_bool],axis=1)/
              np.sum(mu_oldflux[:,lowe_bool],axis=1))
              
mide_bool = (data['ecenters']>3) * (data['ecenters']<30)
change_mid = (np.sum(mu_newflux[:,mide_bool],axis=1)/
              np.sum(mu_oldflux[:,mide_bool],axis=1))

highe_bool = data['ecenters']>30.
change_high = (np.sum(mu_newflux[:,highe_bool],axis=1)/
               np.sum(mu_oldflux[:,highe_bool],axis=1))

plt.plot(data['zcenters'],100*(change_low-1), label='Ev=0.3-3 GeV')
plt.plot(data['zcenters'],100*(change_mid-1), label='Ev=3-30 GeV')
plt.plot(data['zcenters'],100*(change_high-1), label='Ev>30 GeV')

plt.ylim([-0.5,15])
plt.ylabel('Change (%)')
plt.xlabel('cos(theta)')
plt.legend(loc=0)
```

```python
# Modifying 
# a) Nu to NuBar ratio
# b) Up to Horizontal flux
# Play with the parameters below to see how the flux changes!

# Nu to NuBar ratio:
nu_nubar = 1.     # +1 applies the change to neutrinos, 0 to antineutrinos. Anything in between is shared
nubar_ratio = 1.  # Uncertainties assumed between -1 and 1 for the ratio.

# Up to horizontal flux ratio
upvert_ratio = 1. # Uncertainties assumed between -1 and 1 for the ratio.
```

```python
# This is the magic function from oscFit that modifies the weights
reload(sf)

# Neutrino to antineutrino ratios

# NuMu
numu_weightmod = sf.modRatioNuMuBar(mulist, 
                          nu_nubar,    
                          nubar_ratio) 

# NuE
nue_weightmod = sf.modRatioNuEBar(elist, 
                          nu_nubar,    
                          nubar_ratio) # These 4 are the uncorrelated NuE factors. Very little impact.


# Up to horizontal ratio
numu_weightmod *= sf.modRatioUpHor_NuMu(mulist, upvert_ratio)
nue_weightmod  *= sf.modRatioUpHor_NuE(elist, upvert_ratio)

print 'Average Nu/NuBar change (muon): ', numu_weightmod.mean()-1
print 'Average Nu/NuBar change (electron): ', nue_weightmod.mean()-1
```
